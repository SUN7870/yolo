```python
#yaml파일을 학습이 가능하도록 경로 설정
import yaml

data = { 'train' : '/content/Vehicle_Data/train/images/',
         'val' : '/content/Vehicle_Data/valid/images/',
         'test' : '/content/Vehicle_Data/test/images/',
         'names' : ['Ambulance', 'Bus', 'Car', 'Motorcycle', 'Truck'],
         'nc' : 5 }

with open('/content/Vehicle_Data/Vehicle_Data.yaml', 'w') as f:
  yaml.dump(data, f) # data를 YAML로 변환


with open('/content/Vehicle_Data/Vehicle_Data.yaml', 'r') as f:
  vehicle_yaml = yaml.safe_load(f) # YAML파일의 내용을 파이썬 데이터 구조로 변환
  display(vehicle_yaml)
```
